```python
#Mahade_Hasan 
import copy   

def utility(board,depth):
    
    
    for i in range(0,len(board)):
        if board[i][0]==board[i][1] and board[i][1]==board[i][2]:
            if board[i][0]==6:
                return 1
            elif board[i][0]==8:
                return -1
    for i in range(0,len(board)):
        if board[0][i]==board[1][i] and board[1][i]==board[2][i]:
            if board[0][i]==6:
                return 1
            elif board[0][i]==8:
                return -1
    if board[0][0]==board[1][1] and board[1][1]==board[2][2]:
        
        if board[0][0]==6:
            return 1
        elif board[0][0]==8:
            return -1
    if board[0][2]==board[1][1] and board[1][1]==board[2][0]:
        
        if board[0][2]==6:
            return 1
        elif board[0][2]==8:
            return -1
    return 0

def moveleft(board,depth):
    for i in range(0,len(board)):
        for j in range(0,len(board)):
            if board[i][j]==0:
                return True
    return False

def max_algo(board,depth):
    val=-9999
    
    cost=utility(board,depth)
    if cost==1:
        return cost
    if cost==-1:
        return cost
    if moveleft(board,depth)==False:
        return 0
    for i in range(0,len(board)):
        for j in range(0,len(board)):
            if board[i][j]==0:
                board[i][j]=6
                val=max(val,min_algo(board,depth+1))
                board[i][j]=0
    return val
    
def min_algo(board,depth):
    val=+9999
    cost=0
    cost=utility(board,depth)
    if cost==1:
        return cost
    if cost==-1:
        return cost
    if moveleft(board,depth)==False:
        return 0
    for i in range(0,len(board)):
        for j in range(0,len(board)):
            if board[i][j]==0:
                board[i][j]=8
                val=min(val,max_algo(board,depth+1))
                board[i][j]=0
    return val

def Choose_pc_move(board):
   
    
    alpha=-9999
    depth=1
    for i in range(0,len(board)):
        for j in range(0,len(board)):
            if board[i][j]==0:
                
                board[i][j]=6
                val=min_algo(board,depth)
                if val>alpha:
                    alpha=val
                    new_board=copy.deepcopy(board)
                board[i][j]=0
                    
    return new_board
    
def printt(board):
    print(board[0])
    print(board[1])
    print(board[2])
    print("\n")

board=[[0,0,0],
       [0,0,0],
       [0,0,0]]
printt(board)

while True:#Here AI turn 6 an User turn 8
    print("user turn:")
    x,y=input("Enter xy position Like x y : ").split()
    
    x=int(x)
    y=int(y)
    
    board[x][y]=8
    util=utility(board,0)
    if util==1:
        print("AI win")
        printt(board)
        break
    if util==-1:
        print("You win")
        printt(board)
        break
    if moveleft(board,0)==False:
        print("draw")
        printt(board)
        break
        
        
    new_board=Choose_pc_move(board)
    board=copy.deepcopy(new_board)
    
    util=utility(board,0)
    if util==1:
        print("AI win")
        printt(board)
        break
    if util==-1:
        print("You win")
        printt(board)
        break
    if moveleft(board,0)==False:
        print("draw")
        printt(board)
        break
    printt(board)
```
