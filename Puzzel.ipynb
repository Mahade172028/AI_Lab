```python
# I have put a comment on where you need to write the code. 
import copy
import queue

initial_state = [[1, 5, 4],
                 [8, 2, -1],
                 [3, 7, 6]]

final_state = [[-1, 1, 2],
               [3, 4, 5],
               [6, 7, 8]]


class State(object):  # don't change this class
    def __init__(self, matrix):
        self.matrix = matrix
        self.f = 0
        self.g = 0
        self.h = 0

    def __lt__(self, other):
        return self.f < other.f

    def __repr__(self):
        return str(self.matrix[0]) + "\n" + str(self.matrix[1]) + "\n" + str(self.matrix[2])


def create_child(parent_state, move):
    i = 0
    j = 0
    found = False
    matrix = copy.deepcopy(parent_state)
    for i in range(0, 3): 
        for j in range(0, 3):
            if parent_state[i][j] == -1:
                found = True
                break
        if found:
            break

    if move == "up":
        if (i-1) >= 0 and (i-1) < len(parent_state) and j >= 0 and j < len(parent_state):
            
            temp=matrix[i-1][j]
            matrix[i-1][j]=parent_state[i][j]
            matrix[i][j]=temp
            return matrix
       
        else:
            return -1

    if move == "down":
        if (i+1) >= 0 and (i+1) < len(parent_state) and j >= 0 and j < len(parent_state):
            temp=matrix[i+1][j]
            matrix[i+1][j]=parent_state[i][j]
            matrix[i][j]=temp
            return matrix
        
        else:
            return -1

    if move == "left":
        if i >= 0 and i < len(parent_state) and (j-1) >= 0 and (j-1) < len(parent_state):
            temp=matrix[i][j-1]
            matrix[i][j-1]=parent_state[i][j]
            matrix[i][j]=temp
            return matrix
       
        else:
            return -1
    if move == "right":
        if i >= 0 and i < len(parent_state) and (j+1) >= 0 and (j+1) < len(parent_state):
            temp=matrix[i][j+1]
            matrix[i][j+1]=parent_state[i][j]
            matrix[i][j]=temp
            return matrix
        
        else:
            return -1

def heuristic1(current_matrix):
    distance = 0
    for l in range(0,len(current_matrix)):
        for p in range(0,len(final_state)):
            if current_matrix[l][p]!=final_state[l][p] and current_matrix[l][p]!=-1:
                distance+=1
    # write code for calculating heuristic one: misplaced tiles and return the calculated value
    return distance


def heuristic2(current_matrix):
    distance = 0
    for i in range(0,len(current_matrix)):
        for j in range(0,len(current_matrix)):
            a_value=current_matrix[i][j]
            for m in range(0,len(final_state)): 
                      for n in range(0,len(final_state)):
                           if a_value==final_state[m][n] and a_value!=-1:
                                distance+=abs(i-m)+abs(j-n)
    # write code for heuristic two: sum of euclidean distances
    return distance




# step 3: inserting initial state to the open list
def misplace():
    unexplored = queue.PriorityQueue()  # step 1: declaring an open list
    explored = []  # step 2: declaring a closed list

    current_state = State(initial_state)

    current_state.g = 0
    current_state.h = heuristic1(current_state.matrix)
    #print(current_state.h)
    current_state.f = current_state.g + current_state.h
    unexplored.put(current_state, current_state.f)
    flag=0

    # step 4: while open list is not empty
    while not unexplored.empty():
        if flag==1:
            break
        # write code for step 4(a) and 4(b): pop the state with minimum f using unexplored.get()
        current_state=unexplored.get()
        possible_movements = ["up", "left", "down", "right"]
        # complete code for step 4(c)
        for move in possible_movements:
            if flag==1:
                break
            # check if the number of move exceeds 31. If it is, the given state is not considered solvable
            if current_state.g > 31:
                exit(0)
            new_matrix = create_child(current_state.matrix, move)
            
            if new_matrix != -1:
                new_state=State(new_matrix)
                new_state.g=current_state.g+1
                h1=heuristic1(new_state.matrix)
                if(h1==0):
                    flag=1
                    print(new_state.g)
                    break
                    
                new_state.h=h1

                new_state.f=new_state.g+new_state.h
                unexplored.put(new_state,new_state.f)
        explored.append(current_state)

   
           
def manhatten():
    unexplored = queue.PriorityQueue()  # step 1: declaring an open list
    explored = []  # step 2: declaring a closed list

    current_state = State(initial_state)

    current_state.g = 0
    current_state.h = heuristic2(current_state.matrix)
    #print(current_state.h)
    current_state.f = current_state.g + current_state.h
    unexplored.put(current_state, current_state.f)
    flag=0

    # step 4: while open list is not empty
    while not unexplored.empty():
        if flag==1:
            break
        current_state=unexplored.get()
        possible_movements = ["up", "left", "down", "right"]
        # complete code for step 4(c)
        for move in possible_movements:
            if flag==1:
                break
            # check if the number of move exceeds 31. If it is, the given state is not considered solvable
            if current_state.g > 31:
                exit(0)
            new_matrix = create_child(current_state.matrix, move)
            
            if new_matrix != -1:
                new_state=State(new_matrix)
                new_state.g=current_state.g+1
                h1=heuristic2(new_state.matrix)
                if(h1==0):
                    flag=1
                    print(new_state.g)
                    break
                    
                new_state.h=h1

                new_state.f=new_state.g+new_state.h
                unexplored.put(new_state,new_state.f)
        explored.append(current_state)
misplace()
manhatten()
```
